```python
def best_features(X_constant,y): #when you have first column as constant and you dont want that column in final list
    import statsmodels.api as sm
    cols = list(X_constant.columns)
    while len(cols)>1:
        X_constant=X_constant[cols]
        model = sm.GLM(y,X_constant).fit()
        p_max = model.pvalues.nlargest(1)
        feature_max = p_max.index
        if p_max.values>0.05:
            cols.remove(feature_max)
        else:
            break
    return cols
```

### Logistic Regression - ML

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,roc_curve,accuracy_score,f1_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# Train Test Split

from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

best_params = {'C': 10, 'penalty': 'l2', 'solver': 'newton-cg'}
LR = LogisticRegression(**best_params , multi_class='multinomial')
LR.fit(X_train,y_train)

y_train_pred = LR.predict(X_train)
y_train_prob = LR.predict_proba(X_train)[:,1]

print('Train Results')

print('Train - Confusion Matrix\n',confusion_matrix(y_train,y_train_pred))
print()
print('Train - Accuracy Score',accuracy_score(y_train,y_train_pred))
print()
print('Train - AUC',roc_auc_score(y_train,y_train_prob))
print()
print('Train - Classification Report\n',classification_report(y_train,y_train_pred))
print()
print('Train - F1 Score', f1_score(y_train,y_train_pred))
print()
y_test_pred = LR.predict(X_test)
y_test_prob = LR.predict_proba(X_test)[:,1]

print('Test Results' )

print('Test - Confusion Matrix\n',confusion_matrix(y_test,y_test_pred))
print()
print('Test - Accuracy Score',accuracy_score(y_test,y_test_pred))
print()
print('Test - AUC',roc_auc_score(y_test,y_test_prob))
print()
print('Train - Classification Report\n',classification_report(y_test,y_test_pred))
print()
print('Train - F1 Score', f1_score(y_test,y_test_pred))
```

```python
# Cross Vallidation

from sklearn.model_selection import KFold,cross_val_score

kfold= KFold(shuffle=True,n_splits=5,random_state=0)
cv_results = cross_val_score(LR ,X,y, cv=kfold, scoring='f1')  # Scoring can be different according to problem statement
print(cv_results)
print()
print('%s : Mean %f Variance %f' % ('Logistic Regression CV score' , np.mean(cv_results),np.var(cv_results)))
```

```python
# example of grid searching key hyperparametres for logistic regression

from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

# define models and parameters
solvers = ['newton-cg', 'lbfgs','saga']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]

# define grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=LR, param_grid=grid, n_jobs=-1, cv=cv, scoring='f1_weighted',error_score=0)
grid_search.fit(X, y)

# summarize results
print("Best: %f using %s" % (grid_search.best_score_, grid_search.best_params_))
```
